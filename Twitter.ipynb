```python
import string
import operator
import sys
import pdb

FILE_PATH=''
N=0
def inputPath():
    INPUTFILE = input("Input File location ")
    FILE_PATH = INPUTFILE+ '.txt'
    A=input("No. of record: ")
    N=int(A)
    return FILE_PATH,A
    
def MostTweeted(FILE_PATH,A):
    N=int(A)
    with open (FILE_PATH, encoding = "latin-1") as myFile:
        twitter=myFile.readlines()
    l = {}
    for dat in twitter:
       
        fileTemp = dat.split()
        if fileTemp[0] in l:
            l[fileTemp[0]] +=1
        else:
            l[fileTemp[0]] = 1
    l = sorted(l.items(), key = operator.itemgetter(1), reverse = True)
 
    outputFile = open('/Users/pratikkontamwar/MostTweeted.txt', 'w', encoding = "utf-8")
    outputFile.write("The top " + A + " users who tweeted the most: \n")
    for i in range (0,N):
        outputFile.write("The user " + l[i][0] + " tweeted " + str(l[i][1]) + " times" + "\n")
        
    outputFile.close
    
def countofretweet(FILE_PATH,A):
    N=int(A)
    with open (FILE_PATH, encoding = "latin-1") as myFile:
        twit=myFile.readlines()

    g = {}
    for dat in twit:
  
        fileTemp = dat.split()
        #pdb.set_trace()
        #print(len(fileTemp))
        y = len(fileTemp)-2
        tweet = "\""
        for x in range (4, y):
            tweet += fileTemp[x] + " "
        tweet += " ::::;:::: " + fileTemp[0]
  
        if tweet not in g:
            g[tweet] = int(fileTemp[-1])
    #pdb.set_trace()
    outputFile = open('/Users/pratikkontamwar/countofretweet.txt', 'w', encoding = "utf-8")
    g = sorted(g.items(), key = operator.itemgetter(1), reverse = True)
    outputFile.write("The top "+ A + "tweets having max retweet: " + "\n\n")

    for x in range (0, N):
        outputFile.write("\n Tweet: " +
                      g[x][0].split("::::;::::")[0]  + "\n\n")
    outputFile.close    

def Followerstweeted(FILE_PATH,A):
    N=int(A)
    with open (FILE_PATH, encoding = "latin-1") as myFile:
        twitter=myFile.readlines()

    l = {}
    for dat in twitter:
        fileTemp = dat.split()
        if fileTemp[0] not in l:
            l[fileTemp[0]] = int(fileTemp[-2])

    l = sorted(l.items(), key = operator.itemgetter(1), reverse = True)
    outputFile = open('/Users/pratikkontamwar/Followerstweeted.txt', 'w', encoding = "utf-8")
    outputFile.write("The top " + A + "users with maximum followers: " + "\n\n")

    for i in range (0, N):
        outputFile.write(str(i+1) + ". Username: " + l[i][0] + " -> Follower count: " + str(l[i][1]) + "\n\n")
    outputFile.close    
    
def Tweetedperhour(FILE_PATH,A):
    N=int(A)
    with open (FILE_PATH, encoding = "latin-1") as myFile:
        data=myFile.readlines()

    g = {}
    for dat in data:
        fileTemp = dat.split()
        fileTemp2 = fileTemp[1].split(":")
        twitTemp = fileTemp[0] + " " + fileTemp2[1]
        if twitTemp in g:
            g[twitTemp]+=1
        else:
            g[twitTemp]=1
    g = sorted(g.items(), key = operator.itemgetter(1), reverse = True)

    g2={}
    totalUsersIn = 0
    for dat in g:
        totalUsersIn+=1
        if(dat[0].split()[1]) in g2:
            g2[dat[0].split()[1]]+=1
        else:
            g2[dat[0].split()[1]]=1
    g2 = sorted(g2.items(), key = operator.itemgetter(1))

    totalEntriesToPrint = N*len(g2)
    outputFile = open('/Users/pratikkontamwar/Tweetedperhour.txt', 'w', encoding = "utf-8")
   
    outputFile.write("The top " + A + " users who tweeted the most in hour: " + "\n\n")
    for x in range (0,len(g2)):
   
        mSearch = N
        
       
        for dat in g:
            if mSearch == 0:
                break
            if dat[0].split()[1] == g2[x][0]:
                outputFile.write("Username: " + dat[0].split()[0] + "\n\n")
                
                mSearch-=1
    outputFile.close
    


def Followerstweeted(FILE_PATH,A):
    N=int(A)
    with open (FILE_PATH, encoding = "latin-1") as myFile:
        twitter=myFile.readlines()

    l = {}
    for dat in twitter:
        fileTemp = dat.split()
        if fileTemp[0] not in l:
            l[fileTemp[0]] = int(fileTemp[-2])

    l = sorted(l.items(), key = operator.itemgetter(1), reverse = True)
    outputFile = open('/Users/pratikkontamwar/Followerstweeted.txt', 'w', encoding = "utf-8")
    outputFile.write("The top " + A + "users with maximum followers: " + "\n\n")

    for i in range (0, N):
        outputFile.write(str(i+1) + ". Username: " + l[i][0] + " -> Follower count: " + str(l[i][1]) + "\n\n")
    outputFile.close
    
f,a=inputPath()
countofretweet(f,a)
MostTweeted(f,a)
Followerstweeted(f,a)
Tweetedperhour(f,a)
```
